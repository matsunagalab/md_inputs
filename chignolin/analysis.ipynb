```python
import mdtraj as md
import matplotlib.pyplot as plt
```

```python
native = md.load('1uao.pdb')
```

```python
traj = md.load('run.dcd', top='init.psf')
traj = traj.atom_slice(traj.topology.select("protein"))
```

```python
traj.superpose(native)
rmsd = md.rmsd(traj, native) * 10.0
```

```python
plt.plot(rmsd)
plt.xlabel('step')
plt.ylabel('RMSD (Angstrom)')
plt.savefig("analysis.png", dpi=350)
```
